```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec
import glob
```

```python
myfiles = glob.glob(r"C:\Users\Dianne Kenyah\OneDrive - University of Ghana\Documents\GitHub\final_year_worldclim\worldclim\wc2.1_5m_bio\*.tif")
res = []
for i in myfiles:
    curr_file = rasterio.open(i)
    data_bin = curr_file.read()
    data_bin = data_bin.squeeze() != curr_file.nodata
    res.append(data_bin)
bio_data = np.dstack(res)
```
