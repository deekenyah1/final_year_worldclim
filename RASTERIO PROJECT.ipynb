```python
import numpy as np
import rasterio
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


data = dataset1a.read(1)  
nodata = dataset1a.nodata

coords = np.column_stack(np.nonzero(data != nodata))
values = data[data != nodata].reshape(-1, 1)  # Reshape for K-means

kmeans = KMeans(n_clusters=2).fit(values)
labels = kmeans.labels_

output = np.full(data.shape, -1)
output[data != nodata] = labels

plt.figure(figsize=(12, 8))
plt.imshow(output, cmap='plasma')
plt.colorbar()
plt.title('K-means Clustering')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.show()
```

```python
data = dataset1a.read() 
nodata = dataset1a.nodata

coords = np.column_stack(np.nonzero(np.all(data != nodata, axis=0)))  # Get all coordinates where data has values
flattened_data = data[:, coords[:, 0], coords[:, 1]].T  # Reshape for K-means

# Apply K-means
n_clusters = 2  # Number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(flattened_data)
labels = kmeans.labels_

# Plot the results as a scatter map
plt.figure(figsize=(12, 8))
scatter = plt.scatter(coords[:, 1], coords[:, 0], c=labels, cmap='plasma', s=1)
plt.colorbar()
plt.title('K-means Clustering (including no-data values)')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.gca().invert_yaxis()  # Invert y-axis to match the raster's origin
plt.show()
```
